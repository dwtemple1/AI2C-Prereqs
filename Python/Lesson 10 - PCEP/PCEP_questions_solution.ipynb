Part 1: Computer Programming and Python Fundamentals:

```python
# The "Division" Trick
# What is the result of the following code? Why does it return this result?

print(4 / 2)

# The result of the code is 2.0
```

```python
# If the code above returns 2.0, what is the data type of the result?

print(type(4 / 2))

# The result of the code is <class 'float'>
# This is because the division operator (/) in Python always returns a float, even when the division results in a whole number.
# In this case, 4 divided by 2 equals 2, but since the division operator is used, the result is returned as a float (2.0).
```

```python
# Operator Precedence
# What is the result of the following code?

print(2 + 3 * 5 ** 2)
print( (2 + 3) * (5 ** 2) )

# The result of the code varies based on the order of operations.
# This is because of the order of operations in Python, which follows PEMDAS
```

```python
# The "Input" Gotcha
# If a user types 10 when prompted by x = input(), what is the data type of x? Let's find out.

x = input("Enter a number: ")
print(x)
print(type(x))

# The data type of x is <class 'str'>
# This is because the input() function in Python always returns user input as a string.
# If you want to treat the input as a number, you need to convert it using int(input()) or float(input()).
```

```python
# Keyword Sensitivity
# Is True = 5 a valid statement in Python? Why or why not?

print(True == 5)
print(False == 5)
print(True == 1)
print(False == 0)

# The statement True = 5 is not valid in Python because True is a reserved keyword that represents the boolean value for truth.
# However, the comparison True == 5 evaluates to False because True is equivalent to 1
# This is the same case for False, which is equivalent to 0.
```

```python
# Scientific Notation
# What is the value of 3e2?

print(3e2)

# This is just to show that Python can handle scientific notation.
```

Part 2: Control Flow - Conditional Blocks and Loops:

```python
# The "range()"" Stop Value
# How many numbers will range(1, 5) generate?

print(len(range(1, 5)))
```

```python
# Why is that the case? Let's see the numbers generated.

for num in range(1, 5):
    print(num)

# The range(1, 5) function generates numbers starting from 1 up to, but not including, 5.
# Therefore, it generates the numbers 1, 2, 3, and 4
```

```python
# Short-Circuiting Logic
# What is the result of True or (1 / 0)?

print(True or (1 / 0))

# The result of the code is True
# This is because Python sees True and stops evaluating the rest of the expression, so the divide by zero never happens.
```

```python
# Let's try it the other way around

print((1 / 0) or True)

# That looks a little bit better!
```

```python
# Negative Step in Range
# What does range(5, 0, -1) generate?

for num in range(5, 1, -1):
    print(num)

# The range(5, 1, -1) function generates numbers starting from 5 down to, but not including, 1, decrementing by 1 each time.
# This works the same as the normal range function, but it goes backwards because of the negative step value.
```

```python
# The "else" in Loops
# Does the else block in a for loop always execute? Let's find out.

for i in range(3):
    print(i)
else:
    print("Loop completed successfully.")

# Yes, the else block in a for loop always executes after the loop completes normally.
# However, if the loop is terminated prematurely using a break statement, the else block will not execute.

```

```python
# Let's see what happens with a break statement

for i in range(3):
    print(i)
    if i == 1:
        break
else:
    print("Loop completed successfully.")
```

```python
# Identity vs. Equality
# What is the difference between 'is' and '==' in Python? Let's see with an example.

a = [1, 2, 3]

b = a

print(a == b)  # True, because the contents are the same
print(a is b)  # True, because both refer to the same object in memory
```

```python
# Let's try something different

c = [1, 2, 3]

print(a == c)  # True, because the contents are the same
print(a is c)  # False, because they are different objects in memory
```

Part 3: Data Collection (Lists, Tuples, Dicts):

```python
# Negative Slicing
# Given vals = [10, 20, 30, 40, 50], what is vals[-1]?

vals = [10, 20, 30, 40, 50]
print(vals[-1])  # 50
print(vals[-3])  # 30

# How could that be? 
# Negative slicing starts from the end of the list, with -1 referring to the last element.
# Instead of counting from the beginning (0, 1, 2, ...), negative indices count from the end (-1, -2, -3, ...).
```

```python
# List vs. String Slicing
# If s = "Hello, World!", what is s[7:12]?

s = "Hello, World!"
print(s[4:10])  # "World"

# This works the same as list slicing, where the start index is inclusive and the stop index is exclusive.
```

```python
# The "In-Place" Trap
# What is the result of print([3, 1, 2].sort())?

print([3, 1, 2].sort())

# The result of the code is None
# This is because the sort() method sorts the list in place and does not return a new sorted list.
```

```python
# To get the desired result, you need to sort the list first and then print it:

print(sorted([3, 1, 2]))
```

```python
# Tuple Immutability
# Will tup = (1, 2); tup[0] = 0 raise an error? Why or why not?

print(tup = (1, 2))

# tup[0] = 0  # This will raise a TypeError because tuples are immutable.
# Tuples are immutable, meaning their elements cannot be changed after creation.
# Attempting to assign a new value to an element of a tuple raises a TypeError.
```

```python
# Dictionary Keys
# Can a list be used as a key in a dictionary? Why or why not?
# No, a list cannot be used as a key in a dictionary because lists are mutable and unhashable.
# Dictionary keys must be immutable types, such as strings, numbers, or tuples
```

```python
# Strings, Integers, and Tuples are valid keys
my_dict = {
    "name": "Alice",        # String key
    101: "Room Number",     # Integer key
    (1, 2): "Coordinates"   # Tuple key (since tuples are immutable)
}

print(my_dict[(1, 2)]) # Output: Coordinates
```

```python
# This will CRASH
illegal_dict = {
    [1, 2]: "List as a key" 
}

# Error: TypeError: unhashable type: 'list'
```

```python
# Completely valid dictionary
mixed_keys = {
    "status": True,
    5: [1, 2, 3],       # The VALUE is a list (legal)
    (0, 0): "Origin"    # The KEY is a tuple (legal)
}
```

Part 4: Functions and Exceptions:

```python
# Shadowing Scope
# If x = 5 is outside a function and inside the function you have x = 10, what is the value of x inside and outside the function?
# Inside the function, x will be 10 because the local assignment shadows the global variable.
# Outside the function, x will remain 5.

def my_function():
    x = 10
    print("Inside function, x =", x)

x = 5
my_function()

print("Outside function, x =", x)
```

```python
# Default Arguments
# In def func(a, b=2):, what happens if you call func(3)?

def func(a, b=2):
    print("a =", a)
    print("b =", b)

func(3)

# The value of 'a' will be 3 and the value of 'b' will be 2, since 'b' uses its default value.
```

```python
# The "None" Return
# What does a function return if there is no return statement?

def no_return_function():
    print("This function has no return statement.")

result = no_return_function()
print(result)

# The function returns None if there is no return statement.
```

```python
# Exception Hierarchy

try:
    x = 1 / 0
except Exception: # This catches EVERYTHING
    print("Caught by Exception")
except ZeroDivisionError: # This code is UNREACHABLE
    print("Caught by ZeroDivision")

# The code will print "Caught by Exception" because the Exception class is a base class for all built-in exceptions,
# including ZeroDivisionError.
```

```python
# Try-Except-Else

# BLOCK    WHEN DOES IT RUN?
# try      Always runs first. Contains the ""risky"" code
# except   Runs only if an error occurred in the try block
# else     Runs only if NO error occurred in the try block
# finally  Always runs, regardless of whether there was an error or not

try:
    num = 10
    print("Try")
except:
    print("Except")
else:
    print("Else")
finally:
    print("Finally")

# Visualizing the flow:
# 1. "Try" is printed first.
# 2. Since no error occurs, the except block is skipped.
# 3. "Else" is printed next because there was no error.
# 4. "Finally" is printed last, as it always runs.
```

```python
try:
    num = 10 / 0
    print("Try")
except:
    print("Except")
else:
    print("Else")
finally:
    print("Finally")

# Visualizing the flow:
# 1. An error occurs during num = 10 / 0, so "Try" is not printed because the error occurs before it.
# 2. The except block is executed next, printing "Except".
# 3. The else block is skipped because there was an error.
# 4. "Finally" is printed last, as it always runs.
```
